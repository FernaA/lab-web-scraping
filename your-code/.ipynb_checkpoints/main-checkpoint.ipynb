#### Display all the image links from Walt Disney wikipedia page

````python
# This is the url you will scrape in this exercise
url3 = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
#your code

request3 = requests.get(url3).content
soup3 = BeautifulSoup(request3, 'lxml')

name_list3 = []
for a in soup3.find_all('img', src=True):
                        name_list3.append(a['src'])  #a es el indice con el que iteras en soup3, esto mete en losta3 todos los elementos que tienen src.
    
name_list3
````

#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page

````python
# This is the url you will scrape in this exercise
url4 ='https://en.wikipedia.org/wiki/Python' 
````

````python
#your code

request4 = requests.get(url4).content
soup4 = BeautifulSoup(request4, 'lxml')

name_list4 = []
for a in soup4.find_all('a', href=True):
                        name_list4.append(a['href'])  #a es el indice con el que iteras en soup3, esto mete en losta3 todos los elementos que tienen src.
    
name_list4
````
